```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr = [348,450,378,520]
```

```python
mark_arr = np.array(arr)
```

```python
mark_arr
```

```python
df = pd.Series(mark_arr,index=[9,10,11,12])
```

```python
df
```

```python
c=['red','black','green','yellow']
df.plot(kind='bar',color=c)
plt.show()
```

```python
df.plot()
```

```python
plt.scatter(x=[9,10,11,12],y=df)
plt.show()
```

```python
df.plot(kind='pie',legend=True)
plt.show()
```
